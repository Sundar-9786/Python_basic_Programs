```python
# Count of each value

issue_columns = [
    'ISSUE1_INVALID RETAILER STATUS', 'ISSUE2_INVALID RETAILER MODE',
       'ISSUE3_INVALID WHITELIST NUMBER', 'ISSUE4_INVALID VIRTUAL NUMBER',
       'ISSUE5_SIM BLOCKED',  'CREATEDBY',
       'OFFMGRID', 'ACCMGRID', 'HOTSPOTID', 
       'SHOPTYPE',  
       'RETAILERMODE', 'RETAILER_STATUS',  'WHITEIST_STATUS',
       'REGISTRATIONS'
]

for col in issue_columns:
    with pd.option_context('display.max_rows', None):
        
        print(f"\nValue counts for '{col}':")
        print(df[col].value_counts(dropna=False))
```
