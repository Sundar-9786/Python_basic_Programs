### Missing Values of All Data Frames

```python
# Let's assume your DataFrames are named df5, df6, df7, df8, df9

dataframes = {
    'df5': df5,
    'df6': df6,
    'df7': df7,
    'df8': df8,
    'df9': df9
}

# Loop through each column (assuming all dataframes share the same columns)
for column in df9.columns:
    print("=" * 120)
    print(f"{column:^120}")   # nicely centered column name
    print("=" * 120)
    
    for name, df in dataframes.items():
        total_rows = len(df)
        dtype = df[column].dtype
        unique_values = df[column].nunique(dropna=True)
        null_count = df[column].isnull().sum()
        null_percent = (null_count / total_rows) * 100

        # Check if numeric and count zeros
        zero_count = df[column].eq(0).sum() if dtype in ['float64', 'int64'] else "N/A"
        
        print(f"{name:<5} → Rows: {total_rows:<7} | Nulls: {null_count:<6} | "
              f"Missing %: {null_percent:6.2f}% | Dtype: {str(dtype):<8} | "
              f"Unique: {unique_values:<8} | Zero Count: {zero_count}")
    
    print("\n")
```
