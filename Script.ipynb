```python
# Let's assume your DataFrames are named df5, df6, df7, df8, df9

dataframes = {
    'df5': df5,
    'df6': df6,
    'df7': df7,
    'df8': df8,
    'df9': df9
}

# Loop through each column (assuming all dataframes share the same columns)
for column in df9.columns:
    print("=" * 80)
    print(f"{column:^80}")   # nicely centered column name
    print("=" * 80)
    
    for name, df in dataframes.items():
        total_rows = len(df)
        null_count = df[column].isnull().sum()
        null_percent = (null_count / total_rows) * 100
        
        print(f"{name:<5} → Rows: {total_rows:<7} | Nulls: {null_count:<6} | "
              f"Missing %: {null_percent:6.2f}%")
    
    print("\n")
```
